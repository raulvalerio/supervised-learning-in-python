```python
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
import xgboost as xgb
```

```python
data= pd.read_csv("mydatasets/titanic_t.csv")
```

###  Choosing features :  Feature Selection

```python
data= data[['Pclass','Sex','Age','Survived','Parch','SibSp']]
```

```python
data.head()
```

###  Encoding variables

```python
data['Sex'] = data['Sex'].map( {'male':0,'female':1} )
```

```python
data.head(10)
```

### Null variables and missing values

```python
data = data.dropna()
```

### Select output and input variables

```python
X = data.drop('Survived',axis=1)
y = data['Survived']
```

### Split train and test

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,random_state =1)
```

### Chossing Classifier

```python
model = XGBClassifier()
```

### Fitting model

```python
model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

xgb.plot_importance(model, ax=plt.gca())
```

### Predicting new values

```python
predictions = model.predict(X_test)
```

```python
predictions[1:10]
```

```python
X_test.loc[0]
```

## Accuracy

#### Training accuracy

```python
model.score(X_train, y_train) ## accuracy
```

```python
pred_train = model.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train,pred_train)
```

#### Testing accuracy

```python
model.score(X_test, y_test)
```

```python
confusion_matrix(y_test,predictions)
```
